```python
#To solve this issue, we use Imputation which is the process of substituting the missing values in our dataset. 
#A customised function or the Imputer class provided by sklearn can be used for this purpose.

from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values="NaN", strategy='mean')
imputer = imputer.fit(df[['Age']])
df['Age'] = imputer.transform(df[['Age']])
```

```python
df.isnull().sum()
```

```python
df.drop(['Cabin'], axis=1, inplace=False)

#we drop the "Cabin" column since it has a lot of nan values which cannot be imputed 
#and the column does not provide any significant information for this problem
```

```python
#In Standardization we transform our values such that the mean of the values is 0 and the standard deviation is 1.
#fit_transform is equivalent to using fit and then transform.
#fit function calculates the mean and standard deviation and the transform function actually standardizes the dataset.

from sklearn.preprocessing import StandardScaler
std = StandardScaler()
data1 = std.fit_transform(df)
```

```python
#Categorical variables are basically the variables that are discrete and not continuous.
#Two types: Ordinal(ordered) and Nominal(unordered).

#Ordinal Categorical Variables (Example:size, classlabel)
#using map() function:
df_cat = pd.DataFrame(data = [['green','M',10.1,'class1'],
                              ['blue','L',20.1,'class2'],
                              ['white','M',30.1,'class1']])
df_cat.columns = ['color','size','price','classlabel']

size_mapping = {'M':1,'L':2}
df_cat['size'] = df_cat['size'].map(size_mapping)
```

```python
#usinf LabelEncoder() function:
from sklearn.preprocessing import LabelEncoder
class_le = LabelEncoder()
df_cat['classlabel'] = class_le.fit_transform(df_cat['classlabel'].values)
```

```python
print(df_cat)
```

```python
#Nominal Categorical Variables (Example:colour)
#using One-Hot Encoding
#In this technique, we essentially create â€™nâ€™ columns where n is the number of unique values that the nominal variable can take.
#The easiest way to use One-Hot Encoding is to use the get_dummies() function.
#get_dummies function considers only string variables
df_cat = pd.get_dummies(df_cat[['color','size','price']])
```

```python
print(df_cat)
```

```python
#Multicollinearity occurs in our dataset when we have features which are strongly dependent on each other.
#The easiest method to identify Multicollinearity is to just plot a pairplot and you can observe the relationships between 
#different features. If you get a linear relationship between 2 features then they are strongly co-related with each other 
#and there is multicollinearity in your dataset.
#We can use drop_first=True in order to avoid the problem of Multicollinearity.

df_cat = pd.get_dummies(df_cat[['color','size','price']],drop_first=True)
print(df_cat)
```
